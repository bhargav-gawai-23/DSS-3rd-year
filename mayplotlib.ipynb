```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

Line Chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Bar Chart

```python
plt.bar(x,y,color="#23C26F")
plt.title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Scatter Plot

```python
a=(1,2,5,7,6,3,5,4,9,8,5)
b=(21,25,33,56,44,12,21,58,75,89,96,)
plt.scatter(a,b)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Histogram

```python
H=[2,2,5,4,3,3,6,6,8,8,5,5,2,2,1,1,4,4,3,3]
```

```python
plt.hist(H)
plt.show()
```
